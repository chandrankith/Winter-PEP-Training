# Decoraters


In Python, the dataclass decorator is a feature introduced in python 3.7

The dataclass decorator automatically generates the following methods for you:

1. init(): initializes the object snd assigns the provided values to the attributes
2. repr(): Provides a string representation of the object
3. eq(); implements equality comparison between two objects of the class based on their attributes

```python
from dataclasses import dataclass

@dataclass
class Rectangle:
    width: int
    height: int
    color: str = 'white'

rectangle1 = Rectangle(12, 14)
rectangle2 = Rectangle(13, 14, 'red')

print(rectangle1)
print(rectangle2)
```

```python
@dataclass(frozen=True)
class Point:
    x:int
    y:int
```

```python
point=Point(3,4)
```

```python
point.x,point.y
```

## Inheritance

```python
#inheritance

@dataclass
class Person:
    name:str
    age:int

@dataclass
class Employee(Person):
    employee_id:str
    department:str
```

```python
#Nested Dataclasses
@dataclass
class Address:
    street:str
    city:str
    zip_code:str

@dataclass
class Person:
    name:str
    age:str
    address:Address
```

```python
address=Address('123 Main street','Banglore','12345')
person=Person("Krish",31,address)
```

```python
person.address.city
```
