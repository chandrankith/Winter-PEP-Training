# Python OOPS-Public, Protected And Private

```python
## All the class variables are public
class Car():
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
```

```python
audi=Car(4,5,"Diesel")
```

```python
audi.windows=5
```

```python
audi.windows
```

```python
# All the class varaiables are protected
class Car():
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
```

```python
class truck(Car):
    def __init__(self, windows, doors, enginetype,horsepower):
        super().__init__(windows, doors, enginetype)
        self.horsepowwer=horsepower
```

```python
truck._doors=5
```

```python
truck._doors
```

```python
# All the class varaiables are private
class Car():
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
```

```python
audi=Car(4,4,"Diesel")
```

```python
audi._Car_doors=5
```

```python
audi._Car_doors
```
