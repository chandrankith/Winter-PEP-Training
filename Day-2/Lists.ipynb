# List



### Lists are mutable, ordered and can be Heterogeneous


## Range



python tends to follow half-open([start, end])with range and slices:
    *end - start = length
    *easy to concat ranges w/o overlap (ie list(range(3))) + list(range(3,9))




```python
#find range
animals = ["cat", "dog", "bird"]
for index in range(len(animals)):
    print(index, animals[index])
```

## enumerate 
### replaces range


```python
#if you need indices, use ''enumerate'' (to replace ''range(len(list))'')
animals = ["cat", "dog", "bird"]
for index, value in enumerate(animals):
    print(index, value)
```

```python
names = ["john", "goud", "ali", "paji"]
```

```python
enumerate(names)
```

```python
list((i - len(names),n)
    for i, n in enumerate(names))

    
```

## Half-open Interval



Two Properties:

    length = end - start

```python
names[0:3]
```

```python
names[3:]
```

### Shallow copies

```python
#shallow copies
lst1 = [1,2,3,4]
lst2 = lst1.copy()
```

```python
lst2[1] = 1000
```
