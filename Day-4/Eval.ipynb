# How does Eval work

-parse python expression

-compile into a byte code

-Evaluate the python expression

-it will return the result

```python
var = compile("5*5", "<string>", "eval")
```

```python
eval(var)
```

## Globals

```python
## Globals
x = 10
eval("x+59+x**2",{"x":x})
```

```python
x = 100
z = 100
eval("x+z+w", {"x":x, "z":100, "w":1000})
```
