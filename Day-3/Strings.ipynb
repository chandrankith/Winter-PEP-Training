### F-strings

```python
#f-strings
minutes = 36
paragraph = f"""Greeting {name.title()},
Thank you for attending tonight,
we will be here for {minutes/60:.2f} hours
long-winded talk,
Goodbye :{name}!"""
print(paragraph)
```

### String operations

```python
my_str = ('ankith')
print(my_str.isalnum()) #check if all char are numbers
print(my_str.isalpha()) #check if all char in the string are alpha
print(my_str.isdigit()) #test if string contains digits
print(my_str.istitle()) #test if string contains title words
print(my_str.isupper()) #test if string contains upper case
print(my_str.islower()) #test if string contains lower case
print(my_str.isspace()) #test if string contains spaces
print(my_str.endswith('g')) #test if string endswith a d
print(my_str.startswith('a')) #test if string startswith H      
```

### Insert

```python
lst = ["angu", "anshu", "ping", "abs"]
lst.insert(2,"Naik")

print(lst)

```

### Extend Method

```python
lst = [1,2,3,4,5] #can add more compared to append
lst.extend([8,9])
print(lst)
```

### Pop() Method

```python
lst = [1,4,5,6,8,2]
lst.pop()
print(lst)
```
