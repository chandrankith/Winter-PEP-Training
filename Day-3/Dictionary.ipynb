# Dictionary

```python
#can loop over lists, strings,iterators. dictionaries... sequence-like things
my_dict = {"name": "matt", "cash": 5.45}
for key in my_dict: #loop over keys
    print(key)
```

```python
for value in my_dict.values():
    print(value)
```

```python
my_name = {"name": ["ankith", "jatin", "abs"], "numbers": [2, 3, 5] }
for key in my_name:
    print(key)
```

```python
for value in my_name["name"]:
    print(value)
for i in my_name["numbers"]:
    print(i)
```

```python
Bois_dict = {"name": ["abhi", "pallu", "ali", "sanju"]}
for value in Bois_dict.values():
    print(value)
```

Map keys to values, Called associative arrays or hashmaps in other languages.

```python
#literals and constructors
types = {'name': str, 'age': int, 'address': str}
type2 = dict(name=str, age=int, address=str)

#the "key" (''name'') must be hashable
types['name']
```

### Nested Dictionary

```python
car1_model = {'Mercedes': 1960}
car2_model = {'Audi': 1970}
car3_model = {'Bmw': 1980}
car_type= {'car1': car1_model, 'car2': car2_model, 'car3': car3_model}
```

```python
print(car_type)
```

```python
print(car_type['car3'])
```

```python
print(car2_model['Audi'])
```
