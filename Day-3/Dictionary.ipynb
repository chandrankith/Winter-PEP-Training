# Dictionary

```python
#can loop over lists, strings,iterators. dictionaries... sequence-like things
my_dict = {"name": "matt", "cash": 5.45}
for key in my_dict: #loop over keys
    print(key)
```

```python
for value in my_dict.values():
    print(value)
```

```python
my_name = {"name": ["ankith", "jatin", "abs"], "numbers": [2, 3, 5] }
for key in my_name:
    print(key)
```

```python
for value in my_name["name"]:
    print(value)
for i in my_name["numbers"]:
    print(i)
```

```python
Bois_dict = {"name": ["abhi", "pallu", "ali", "sanju"]}
for value in Bois_dict.values():
    print(value)
```
